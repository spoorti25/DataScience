--> start variable name by alphabets only
--> use the underscore to separate the words
--> use number after the word

```python
var = 10
var1 = 20
var_num = 30
```

```python
--> any special char/ number should not be used 1st.
--> no spaces used between the word
--> variable name should not be the keywords
--> case sensitive [a, A] different.
```

## Calculate the 95% confidence for our mean point estimate

```python
np.random.seed(10)
print("population mean estimate:",population_ages.mean())

pop_stdev = population_ages.std()  # Get the population standard deviation

sample_size = 5000
sample = np.random.choice(a= population_ages, size = sample_size)
sample_mean = sample.mean()

print('sample_mean',sample_mean)
print() 
z_critical = stats.norm.cdf(x = 0.95)  # Get the z-critical value*

print("z-critical value:",z_critical)              # Check the z-critical value
print()                        

margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)  
print("Confidence interval:",confidence_interval)
print()
```

## Chisquares test

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
national = pd.DataFrame(["white"]*100000 + ["hispanic"]*60000 +\
                        ["black"]*50000 + ["asian"]*15000 + ["other"]*35000)
minnesota = pd.DataFrame(["white"]*600 + ["hispanic"]*300 + \
                         ["black"]*250 +["asian"]*75 + ["other"]*150)

print(national)

national_table = pd.crosstab(index=national[0], columns="count")
minnesota_table = pd.crosstab(index=minnesota[0], columns="count")

print( "National")
print(national_table)
print(" ")
print( "Minnesota")
print(minnesota_table)
```

```python
national_ratios = national_table/len(national)  # Get population ratios

observed = minnesota_table
expected = national_ratios * len(minnesota)   # Get expected counts

# chi_square statistics
chi_squared_stat = (((observed-expected)**2)/expected).sum()
print(chi_squared_stat)

crit = stats.chi2.ppf(q = 0.95, # Find the critical value for 95% confidence*
                      df = 4)   # Df = number of variable categories - 1
print("Critical value:", crit)
print()

p_value = 1 - stats.chi2.cdf(x=chi_squared_stat,  # Find the p-value
                             df=4)
print("P value:",p_value)
print()
```
